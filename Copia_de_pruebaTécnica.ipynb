```python
import requests
response = requests.get("https://inspirehep.net/api/institutions?q=colombia")
```

```python
data = response.json()
data
```

```python
#2
datahits = data['hits']
datahits2 = datahits['hits']
print("lista de instituciones de Colombia:")
legacy_ICN_List = []
for key in datahits2:
  metadato = key['metadata']
  if(metadato['number_of_papers'] > 0):
    print(metadato["legacy_ICN"])
    legacy_ICN_List.append(metadato["legacy_ICN"])
```

```python
#3
authors_urls = []
for key in legacy_ICN_List:
  response2 = requests.get("https://inspirehep.net/api/literature?sort=mostrecent&page=1&q=aff+"+key+"+and+ac+1->+10")
  data2 = response2.json()
  for key in data2['hits']['hits']:
    if('record' in key['metadata']['authors'][0]):
      authors_urls.append(key['metadata']['authors'][0]['record']['$ref'])
```

```python
authors_urls
```

```python
#4
listaAutores = []

for url in authors_urls:
  datosAutor = []
  response3 = requests.get(url)
  author = response3.json()
  datosAutor.append(author['metadata']['name']['value'])
  if('email_addresses' not in author['metadata']): datosAutor.append("N/A")
  else: datosAutor.append(author['metadata']['email_addresses'][0]['value'])
  if('positions' in author['metadata']):
    if('rank' not in author['metadata']['positions'][0]): datosAutor.append("N/A")
    else: datosAutor.append(author['metadata']['positions'][0]['rank'])
    if('institution' not in author['metadata']['positions'][0]): datosAutor.append("N/A")
    else: datosAutor.append(author['metadata']['positions'][0]['institution'])
    if('start_date' not in author['metadata']['positions'][0]): datosAutor.append("N/A")
    else: datosAutor.append(author['metadata']['positions'][0]['start_date'])
    if('end_date' not in author['metadata']['positions'][0]): datosAutor.append("N/A")
    else: datosAutor.append(author['metadata']['positions'][0]['end_date'])
    listaAutores.append(datosAutor)
  else:
      datosAutor.append("N/A")
      datosAutor.append("N/A")
      datosAutor.append("N/A")
      datosAutor.append("N/A")
```
